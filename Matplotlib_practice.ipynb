```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
print(heart_disease.head())
len(heart_disease)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
plt.style.use('seaborn')

# plotting figure and axisies
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 6),
                               sharex=True)

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

ax0.set(title="Cholesterol VS Age",
        ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")



scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

ax1.set(title="Max heart rate VS Age",
        xlabel="Age",
        ylabel="Max heart rate")

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")

fig.suptitle("Heart Disease Analyris", fontsize=14);
```

```python
fig.savefig("Heart_disease")
```
